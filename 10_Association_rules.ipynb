```python
#!pip install pandas mlxtend openpyxl
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
df = pd.read_excel("Online retail.xlsx", header=None)
df.head()
```

```python
df.columns
```

```python
df.info()
```

```python
# Data Preprocessing
df.drop_duplicates(inplace=True)
df.dropna(inplace=True)
```

```python
# Convert each row to a list of items (assuming each row is a transaction)
transactions = df[0].apply(lambda x: [item.strip() for item in str(x).split(',')])
```

```python
transactions
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(transactions).transform(transactions)
```

```python
df_encoded = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
df_encoded
```

```python
# Apply Apriori Algorithm
frequent_itemsets = apriori(df_encoded, min_support=0.02, use_colnames=True)
```

```python
rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1.2)
```

```python
# Filter rules for meaningful patterns
rules = rules[(rules['confidence'] >= 0.3) & (rules['lift'] >= 1.2)]
```

```python
# Display top rules
print(rules[['antecedents', 'consequents', 'support', 'confidence', 'lift']].head())
```

```python
# Interpretation Example
for idx, row in rules.iterrows():
    print(f"Rule: {set(row['antecedents'])} -> {set(row['consequents'])} | Lift: {row['lift']:.2f}")
```

```python
rules.head()
```

```python
print("Number of Rules Found:", len(rules))
print("Top Rule:")
if not rules.empty:
    top_rule = rules.sort_values(by='lift', ascending=False).iloc[0]
    print(f"{set(top_rule['antecedents'])} → {set(top_rule['consequents'])}")
    print(f"Support: {top_rule['support']:.2f}, Confidence: {top_rule['confidence']:.2f}, Lift: {top_rule['lift']:.2f}")
else:
    print("No rules found with the given thresholds.")
```

- **Frequent Product Combinations:** The Apriori algorithm identified sets of products that are often purchased together in the same transaction. These combinations represent strong associations in customer buying behavior.

- **High Confidence and Lift Rules:** By filtering for rules with confidence ≥ 0.3 and lift ≥ 1.2, you focused on relationships where the purchase of one product significantly increases the likelihood of another product being bought. High lift values indicate that these associations are much stronger than random chance.

- **Actionable Customer Patterns:** The discovered rules provide actionable insights, such as which products to bundle, promote together, or place near each other in-store to increase sales.

- **Top Rule Example:** The code displays the top rule (highest lift), showing the most influential product association in your dataset, along with its support, confidence, and lift values.

These insights can help optimize product placement, cross-selling strategies, and targeted marketing based on actual customer purchasing patterns.
